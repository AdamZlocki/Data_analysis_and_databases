```python
import pandas as pd
from pathlib import Path  
data = pd.read_csv("../Original_Data/tb.csv")
data
```

```python
data_without_null = data.dropna(subset=['iso2', 'year'])
data_without_null = data_without_null.dropna(axis=0, thresh=4)
data_without_null.drop(['new_sp_m04', 'new_sp_f04','new_sp_m514', 'new_sp_f514','new_sp_mu','new_sp_fu'], inplace=True, axis=1)
data_without_null
```

```python
df = data_without_null.fillna(0)
df = pd.melt(df, id_vars=['iso2', 'year'], value_vars=list(df.columns)[3:],
             var_name='column', value_name='cases')
df['cases'] = df['cases'].astype(int)
df['sex'] = df['column'].str[7]
df['age'] = df['column'].str[8:].map({
    '014': '0-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65': '65+'
})
df = df[['iso2', 'year', 'sex', 'age', 'cases']]
df = df.sort_values(by=['iso2','year', 'age','sex'])

df.rename(columns={"iso2": "country"}, inplace = True)
df.reset_index(drop=True)
```
