```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

PATH = "../Analysis_Data/"

original_data = pd.read_csv("../Original_Data/6_Mazowieckie.csv")
original_data.drop(columns = original_data.columns[0], axis = 1, inplace= True)
original_data
```

```python
data_with_proper_age = original_data
data_with_proper_age= data_with_proper_age.dropna(axis="index", how="any", subset=['Wiek kupującego'])
data_with_proper_age.reset_index(drop=True)
data_with_proper_age.to_csv(f"{PATH}data_with_proper_age_values.csv")
```

```python
data_with_proper_gender = data_with_proper_age
data_with_proper_gender["Płeć kupującego"].replace("bd.", np.nan, inplace=True)
data_with_proper_gender= data_with_proper_gender.dropna(axis="index", how="any", subset=['Płeć kupującego'])
data_with_proper_gender.reset_index(drop=True)
data_with_proper_gender.to_csv(f"{PATH}data_with_specified_in_gender.csv")
```
