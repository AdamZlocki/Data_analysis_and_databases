## 10. Jakie kategorie filmów zostały wypożyczone przez klientów?

```python
txt = """
        SELECT distinct name
        FROM category c
        INNER JOIN film_category fc ON c.category_id = fc.category_id
        INNER JOIN film f ON f.film_id = fc.film_id
        INNER JOIN inventory inv ON inv.film_id = f.film_id
        INNER JOIN rental re ON inv.inventory_id = re.inventory_id
        """
df = pd.read_sql(txt,con=connection)
df
```

## 11. Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce.


```python
txt = """
        SELECT distinct name
        FROM category c
        INNER JOIN film_category fc ON c.category_id = fc.category_id
        INNER JOIN film f ON f.film_id = fc.film_id
        INNER JOIN inventory inv ON inv.film_id = f.film_id
        INNER JOIN rental re ON inv.inventory_id = re.inventory_id
        INNER JOIN customer cu ON re.customer_id = cu.customer_id
        INNER JOIN address ad ON cu.address_id = ad.address_id
        INNER JOIN city ON ad.city_id = city.city_id
        INNER JOIN country cou ON cou.country_id = city.country_id
        WHERE country = 'United States'
        """
df = pd.read_sql(txt,con=connection)
df
# lub

# txt = """
#         SELECT distinct name
#         FROM category c
#         INNER JOIN film_category fc ON c.category_id = fc.category_id
#         INNER JOIN film f ON f.film_id = fc.film_id
#         INNER JOIN inventory inv ON inv.film_id = f.film_id
#         INNER JOIN rental re ON inv.inventory_id = re.inventory_id
#         INNER JOIN staff st ON re.staff_id = st.staff_id
#         INNER JOIN store str ON st.store_id = str.store_id
#         INNER JOIN address ad ON str.address_id = ad.address_id
#         INNER JOIN city ON ad.city_id = city.city_id
#         INNER JOIN country cou ON cou.country_id = city.country_id
#         WHERE country = 'United States'
#         """
# df = pd.read_sql(txt,con=connection)
# df
```

## 12. Znajdź wszystkie tytuły filmów, w których grał: Olympia Pfeiffer lub Julia Zellweger lub Ellen Presley

```python
txt = """
        SELECT distinct title
        FROM film f
        INNER JOIN film_actor f_ac ON f.film_id = f_ac.film_id
        INNER JOIN actor a ON a.actor_id = f_ac.actor_id
        WHERE (first_name = 'Olympia' and last_name = 'Pfeiffer') or (first_name = 'Julia' and last_name = 'Zellweger') or (first_name = 'Ellen' and last_name = 'Presley') 
        """
df = pd.read_sql(txt,con=connection)
df
```
